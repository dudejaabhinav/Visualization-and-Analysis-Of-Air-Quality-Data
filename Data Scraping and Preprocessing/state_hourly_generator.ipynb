```python
import pandas as pd
```

```python
df = pd.read_csv('site_data_hourly.csv')
df
```

```python
df1 = pd.read_csv('site_to_location.csv')
df1
```

```python
merged_df = df.merge(df1[['site_id', 'state']], on='site_id', how='left')
merged_df
```

```python
new_df = merged_df.groupby([merged_df['state'], merged_df['From Date']])[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'SO2', 'CO', 'Ozone', 'AQI']].agg(['min', 'max', 'mean'])
new_df
```

```python
new_df.columns = new_df.columns.map('_'.join)
new_df.index = new_df.index.set_names(['State', 'Date'])
new_df.reset_index(inplace=True)
new_df
```

```python
new_df.to_csv('state_data_hourly.csv', index=False)
```
