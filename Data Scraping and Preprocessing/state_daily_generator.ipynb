import pandas as pd

df = pd.read_csv('site_data_daily.csv')
df

df1 = pd.read_csv('site_to_location.csv')
df1

merged_df = df.merge(df1[['site_id', 'state']], on='site_id', how='left')
merged_df

```python
merged_df['Time'] = merged_df['Time'].replace({'All Day': 'a', 'Night': 'b', 'Morning': 'c', 'Afternoon': 'd', 'Evening': 'e'})
agg_df = merged_df.groupby(['state', 'Date', 'Time']).agg({'PM2.5_min': 'min', 'PM2.5_max': 'max', 'PM2.5_mean': 'mean', 'PM10_min': 'min', 'PM10_max': 'max', 'PM10_mean': 'mean', 'NO_min': 'min', 'NO_max': 'max', 'NO_mean': 'mean', 'NO2_min': 'min',
           'NO2_max': 'max', 'NO2_mean': 'mean', 'NOx_min': 'min', 'NOx_max': 'max', 'NOx_mean': 'mean', 'NH3_min': 'min',
           'NH3_max': 'max', 'NH3_mean': 'mean', 'SO2_min': 'min', 'SO2_max': 'max', 'SO2_mean': 'mean', 'CO_min': 'min',
           'CO_max': 'max', 'CO_mean': 'mean', 'Ozone_min': 'min', 'Ozone_max': 'max', 'Ozone_mean': 'mean', 'AQI_min': 'min',
           'AQI_max': 'max', 'AQI_mean': 'mean'}).reset_index()
agg_df['Time'] = agg_df['Time'].replace({'a': 'All Day', 'b': 'Night', 'c': 'Morning', 'd': 'Afternoon', 'e': 'Evening'})
agg_df
```
